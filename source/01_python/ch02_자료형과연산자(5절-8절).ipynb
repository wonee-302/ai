## 5.3 논리연산자 : and, &, or, |, not

```python
# & : 논리연산자, 비트연산자
print( (10>3)  &  (10>5) ) # 논리연산 True & True => True
print( 3 & 10 ) # 비트연산 수&수
#  3: 0 0 1 1
# 10: 1 0 1 0
# & : 0 0 1 0 (2)
print( 10>3 & 10>5) # 10>2>5
```

```python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 참이면 마지막항을 반환
print( True  and  True and 'abc')
print(0 and 5)
```

```python
# | : 논리연산자, 비트연산자
print( (10>3)  |  (10>5) ) # 논리연산 True | True => True
print( 3 | 10 ) # 비트연산 수&수
#  3: 0 0 1 1
# 10: 1 0 1 0
# | : 1 0 1 1 (11)
print( 10>3 | 10>5) # 10>11>5
```

```python
# or : 논리연산자, 참으로 판별되는 첫번째 항의 결과 반환
#       모든 항이 거짓이면 마지막항을 반환
print( False  or  False or 0)
print(0 or '홍길동')
```
